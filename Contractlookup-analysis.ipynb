```python
word_list = ['smoke','edible','marijuana','cannabis','shop','natural','organic','healing','cbd','relax',
            'high','guns','weapons','firearms','arms','range','shoot','magnum','dispensary']

pat = '|'.join(r"\b{}\b".format(x) for x in word_list)
df1 = (df['Legal_Business_Name'].str.lower()
                            .str.extractall('(' + pat + ')', flags = re.I)[0]
                            .unstack().add_prefix('Word_'))

df = df.join(df1).fillna('')

print(df)
```

```python
df.to_csv('df.csv', index = False)
```
